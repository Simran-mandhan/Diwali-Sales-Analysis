# Exploratory Data Analysis

## Gender

```python
df.columns
```

```python
sns.countplot(x = 'Gender', data = df)
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers :
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

bar = sns.barplot(x = 'Gender' , y = 'Amount' , data = sales_gen)

for bars in bar.containers:
    bar.bar_label(bars)
```

## Age

```python
ax1 = sns.countplot(data = df, x ='Age Group', hue = 'Gender')

for bars in ax1.containers:
    ax1.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize':(18,5)})
bar1 = sns.barplot(x = 'Age Group', y= 'Amount', data = sales_age)

for bars in bar1.containers:
    bar1.bar_label(bars)
 
```

## State

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(18,5)})

bar2 = sns.barplot(x = 'State', y = 'Orders' , data = sales_state)

for bars in bar2.containers:
    bar2.bar_label(bars)
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize': (18,5)})

bar2 = sns.barplot(x = 'State', y = 'Amount', data = sales_state)
```

## Marital Status

```python
ax2 = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc = {'figure.figsize': (7,5)})
for bars in ax2.containers:
    ax2.bar_label(bars)
```

```python
sales_marital = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize': (6,5)})

sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_marital, hue = 'Gender')
```

## Occupation

```python
sns.set(rc = {'figure.figsize': (20,5)})
ax3 = sns.countplot(data = df, x = 'Occupation')

for bars in ax3.containers:
    ax3.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize': (20,5)})
sns.barplot(data = sales_occupation, x = 'Occupation', y = 'Amount')
```

## Product Category

```python
sns.set(rc = {'figure.figsize': (25,6)})
ax4 = sns.countplot(data = df, x = 'Product_Category')

for bars in ax4.containers:
    ax4.bar_label(bars)
```

```python
sales_product = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize': (20,5)})

sns.barplot(data = sales_product, x = 'Product_Category', y = 'Amount')
```

```python
sales_product = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize': (20,5)})

bar3 = sns.barplot(x = 'Product_ID', y = 'Orders', data = sales_product)

for bars in bar3.containers:
    bar3.bar_label(bars)
```
